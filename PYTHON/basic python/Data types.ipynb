# indexed 
# mutable

list = [] # empty list

```python
list = [10,20,30,40,"hello",10.5,100,10] # hetrogenious 
print(list,"ordered same as you insert")
print(list[2])  # indexed 
list[4] = "world" # mutable -- (changable)
print(list)
print(type(list))
```

```python
# tuple -- tuple is a collection of elements ..
# denoted by () parenthesis with csv..
# ordered 
# indexed 
# immutable 
# heteregenious in nature. 
# allows duplicates 
t = ()  # empty tuple 
print(t)
print(type(t))
```

```python
t = (10,) # without comma it is not tuple its through int..
print(t)
print(type(t))
```

```python
t = (10,20,30,"hello",10.5,10) # hetrogenious 
print(t,"ordered") # ordered as you you have inseted
print(type(t))
print(t[2]) # indexed 
t[3] = "world" # immutable (no changable)
print(t)


```

```python
# set --> set is the collections of elements ..
# denoted by {} curly braces with csv .. 
# stored unique value only 
# unordered 
# unindexed 
# mutable 
# heterogenious in nature .. 

s = set() # set constructor (empty set) .. without constructor it takes dictionary ,,, 
print(s)
print(type(s))
```

```python
s = {10,20,40,"hello",10.5,10,10} # stored unique value only 
print(s) # unordered 
s.add(1000) # mutable 
print(s)
#s[2] = 50 # unindexed 
print(s) # heterogenious 
```

```python
# dictionary --- > dictionary is the collections of elements it stored (key value pair)
# denoted by {k:v}
# key must be unoque value can be duplicates
# dictionary do not indexed but we can access the elemets through the (KEY)
# ordered 
# mutable 

d = {} # empty dictionary 
print(d)
print(type(d))
```

```python
d = {"name":"samshad","course":"Ds","fee":"30000"} # ordered as we can inserted 
print(d)
print(d["name"]) # unindexed but access elements through the key .. 
d["name"] = "ansari" # mutable 
print(d)
```

```python
# string -- string is the collections of characters stored value inside the (singel quotes and double quotes and triple quotes).
# ordered 
# indexed 
# immutable 
# homogenious --> only it takes text value . 
# allows duplicates .. 

s = "" # empty string 
print(s)
print(type(s))
```

```python
s = "samshad123" # ordered 
print(s)
print(s[2]) # indexed  
s[2] = "mm" # immutable 
print(s)
```
