```python
# 10. Check if an element exists in a list..
l = [1,2,3,2,4,23,10]
n = int(input("enter any number"))

print(l)
if n in l:
    print(n,"number exist ")
else:
    print(n,"number not exist")
    
```

```python
# 11. Count how many times an element appears in a list.
num = int(input("enter any number"))
l = [1,2,3,4,5,6,10,10,10,3,2,1,100]
count = 0

for i in l:
    if i == num:
        count = count+1 
if count>0:
    print(num,"appear",count,"times")
else:
    print(num,"not exist")
        
```
