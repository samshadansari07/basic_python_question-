```python
# print square .. 
n = int(input("enter number"))
for i in range(n):
    for j in range(n):
        print("*",end = " ")
    print()
```

```python
# print right alinged right angle triangle ...
n = int(input("enter number"))
for i in range(n):
    for j in range(i+1):
        print("*",end = " ")
    print()
```

```python
# inverted right angle triangle .. 
n = int(input("enter number"))
for i in range(n):
    for j in range(n-i):
        print("*",end = " ")
    print()
```

```python
n = int(input("enter number"))
for i in range(n):
    for j in range(n-i-1):
        print("$",end=" ")
    for k in range(i+1):
        print("*",end = " ")
    print()
```

```python
n = int(input("enter any number"))
for i in range(n):
    for j in range(n-i-1):
        print(" ",end=" ")
    for k in range(i+1):
        print("*",end = " ")
    print()
```

```python
n = int(input("enter any number"))
for i in range(n,0,-1):
    for j in range(n-i):
        print(" ",end =" ")
    for k in range(i):
        print("*",end =" ")
    print()
```
