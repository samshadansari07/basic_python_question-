```python
# 12. Reverse a number.
n = int(input("enter any number"))
rev = 0 
while n>0:
    digit = n%10
    rev = rev*10+digit
    n = n//10
print("reverse number :",rev)   
```

```python
# 13. Check palindrome number .
n = int(input("enter any number"))
original = n
num = 0
while n>0:
    digit = n%10
    num = num*10+digit 
    n = n//10
if original==num:
    print("palindrome",original)
else:
    print("not palindrome",original)
```

```python
# 14. Find sum of digits. 
n = abs(int(input("enter any digit ")))
total = 0

while n>0:
    digit = n%10
    total = total+digit
    n = n//10
print("sum of digit",total)

```

```python
# 15. Find product of digits..
n = int(input("enter any number"))

if n == 0:
    print("product of n :",n)
else:
    product = 1
    while n>0:
        digit = n%10
        product = product * digit
        n = n//10
    print("product of number:",product)
```

```python
# 16. Find largest digit in a number.. 
n =int(input("enter any number"))

larghest = 0
while n>0:
    digit = n %10 
    if digit > larghest:
        larghest = digit 
    n = n//10
print("larghest digit:",larghest)
        
```

```python
# 17. Find smallest digit in a number .. 
n = int(input("enter any number"))
smallest = 9 
while n>0:
    digit = n%10
    if smallest>digit:
        smallest = digit
    n = n//10
print("smallest digit:",smallest)
```

```python
# 47. Check if number is prime .. 
num = int(input("enter any number"))
if num <2:
    print(num,"not a prime number")
else:
    for i in range(2,num):
        if num%i== 0 :
            print(num,"not a prime number")
            break
    else:
        print(num,"prime number")
```

```python
# 48. Print all primes till n .. 
num = int(input("enter any number: "))
count = 0

for n in range(2, num + 1):
    for i in range(2, n):
        if n % i == 0:
            break
    else:
        print(n, end=" ")
        count += 1
print()
print(count)
```

```python
#50  count prime number till n .. 
n = int(input("enter any number: "))
count = 0

if n == 0 or n== 1:
    print(n,"it is not a prime number")
elif n>1:
    for num in range(2,n+1):
        for i in range(2,num):
            if num%i == 0:
                break 
        else:
            print(num,"is a prime number")
            count += 1
else:
    print(n,"is invalid")
print("count of prime number till",n,"----> ",count)
```
